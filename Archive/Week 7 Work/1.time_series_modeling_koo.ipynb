```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df =pd.read_csv('final_df_updated.csv')
```

```python
df["Electric Bike Count"] = df["Electric Bike Count"].interpolate()
```

```python
# Define train-test split
split_point = int(len(df) * 0.8)  # 80% train, 20% test
train_df = df.iloc[:split_point]  # First 80% for training
test_df = df.iloc[split_point:]  # Last 20% for testing

# Ensure Date column is in datetime format
train_df['Date'] = pd.to_datetime(train_df['Date'])
test_df['Date'] = pd.to_datetime(test_df['Date'])

# Check shapes
print(f"Train size: {train_df.shape}, Test size: {test_df.shape}")
```

```python
# Plot train and test data
plt.figure(figsize=(10, 5))
plt.plot(train_df['Date'], train_df["Electric Bike Count"], marker="o", linestyle="-", label="Train Data")
plt.plot(test_df['Date'], test_df["Electric Bike Count"], marker="o", linestyle="-", label="Test Data", color="red")
plt.title("Train & Test Split for Electric Bike Count")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# Train ARIMA model
arima_model = ARIMA(train_df["Electric Bike Count"], order=(2, 1, 2))
arima_result = arima_model.fit()

# Forecast future values (same length as test set)
arima_forecast = arima_result.forecast(steps=len(test_df))

# Convert to DataFrame for comparison
arima_forecast_df = pd.DataFrame({"Date": test_df.index, "ARIMA Forecast": arima_forecast})
arima_forecast_df.set_index("Date", inplace=True)

# Plot ARIMA results
plt.figure(figsize=(10, 5))
plt.plot(train_df.index, train_df["Electric Bike Count"], label="Train Data", color="blue")
plt.plot(test_df.index, test_df["Electric Bike Count"], label="Test Data", color="red")
plt.plot(arima_forecast_df.index, arima_forecast_df["ARIMA Forecast"], label="ARIMA Forecast", color="green")
plt.title("ARIMA Forecast for Electric Bike Count")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.show()

# Prepare Prophet dataframe
prophet_train = train_df.reset_index().rename(columns={"Date": "ds", "Electric Bike Count": "y"})
prophet_test = test_df.reset_index().rename(columns={"Date": "ds", "Electric Bike Count": "y"})

# Train Prophet model
prophet_model = Prophet()
prophet_model.fit(prophet_train)

# Create future dates for prediction (using actual test dates)
future_dates = pd.DataFrame({"ds": test_df['Date']})
prophet_forecast = prophet_model.predict(future_dates)

# Ensure datetime format consistency
prophet_forecast["ds"] = pd.to_datetime(prophet_forecast["ds"])

# Merge actual vs predicted values for comparison
merged_df = prophet_test.merge(prophet_forecast[['ds', 'yhat']], on='ds', how="left")

# Extract only relevant forecasted values
prophet_forecast_df = prophet_forecast[["ds", "yhat"]].set_index("ds")

# Plot Prophet results
plt.figure(figsize=(10, 5))
plt.plot(train_df['Date'], train_df["Electric Bike Count"], label="Train Data", color="blue")
plt.plot(test_df['Date'], test_df["Electric Bike Count"], label="Test Data", color="red")
plt.plot(prophet_forecast_df.index, prophet_forecast_df["yhat"], label="Prophet Forecast", color="green")
plt.title("Prophet Forecast for Electric Bike Count")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.show()

# Function to calculate MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# ARIMA Evaluation
arima_mae = mean_absolute_error(test_df["Electric Bike Count"], arima_forecast)
arima_rmse = np.sqrt(mean_squared_error(test_df["Electric Bike Count"], arima_forecast))
arima_mape = mean_absolute_percentage_error(test_df["Electric Bike Count"], arima_forecast)

# Prophet Evaluation
prophet_mae = mean_absolute_error(test_df["Electric Bike Count"], prophet_forecast_df["yhat"])
prophet_rmse = np.sqrt(mean_squared_error(test_df["Electric Bike Count"], prophet_forecast_df["yhat"]))
prophet_mape = mean_absolute_percentage_error(test_df["Electric Bike Count"], prophet_forecast_df["yhat"])

# Print results
print(f"ARIMA MAE: {arima_mae:.2f}, ARIMA RMSE: {arima_rmse:.2f}, ARIMA MAPE: {arima_mape:.2f}%")
print(f"Prophet MAE: {prophet_mae:.2f}, Prophet RMSE: {prophet_rmse:.2f}, Prophet MAPE: {prophet_mape:.2f}%")
```

```python

# Plot train and test data
plt.figure(figsize=(10, 5))
plt.plot(train_df['Date'], train_df["Member User Count"], marker="o", linestyle="-", label="Train Data")
plt.plot(test_df['Date'], test_df["Member User Count"], marker="o", linestyle="-", label="Test Data", color="red")
plt.title("Train & Test Split for Member User Count")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# Dictionary to store evaluation results
evaluation_results = {}

# Iterate over each membership type
for membership_type in ["Casual User Count", "Member User Count"]:
    # Rename columns for Prophet
    train_data = train_df[['Date', membership_type]].rename(columns={'Date': 'ds', membership_type: 'y'})
    test_data = test_df[['Date', membership_type]].rename(columns={'Date': 'ds', membership_type: 'y'})

    # Initialize Prophet model
    model = Prophet()
    model.fit(train_data)  # Train on training set

    # Create future dataframe to predict on test period
    future_dates = pd.DataFrame(test_data['ds'])  # Use actual test dates
    forecast = model.predict(future_dates)  # Forecast for test set

    # Merge actual vs predicted values for comparison
    merged_df = test_data.merge(forecast[['ds', 'yhat']], on='ds')

    # Compute error metrics
    def evaluate_forecast(actual, predicted):
        mae = mean_absolute_error(actual, predicted)
        rmse = np.sqrt(mean_squared_error(actual, predicted))
        mape = np.mean(np.abs((actual - predicted) / actual)) * 100  # Convert to percentage
        return {'MAE': mae, 'RMSE': rmse, 'MAPE': mape}

    errors = evaluate_forecast(merged_df['y'], merged_df['yhat'])
    evaluation_results[membership_type] = errors

    # Plot forecast vs. actual
    model.plot(forecast)
    plt.scatter(test_data['ds'], test_data['y'], color='red', label="Actual Test Data")
    plt.title(f'Prophet Forecast for {membership_type}')
    plt.legend()
    plt.show()

# Print Evaluation Results
for membership_type, metrics in evaluation_results.items():
    print(f"Evaluation for {membership_type}:")
    print(f"MAE: {metrics['MAE']:.2f}")
    print(f"RMSE: {metrics['RMSE']:.2f}")
    print(f"MAPE: {metrics['MAPE']:.2f}%\n")
```
