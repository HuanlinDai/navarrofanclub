```python
# Handling file path for quarterly reports up until 2016 Q3 - 2020 Q1
years = ['2016', '2017', '2018', '2019', '2020']
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
quarter_file_paths = []
for year in years:
    if year == '2016':
        for quarter in quarters[3:5]:
            quarter_file_paths.append(f'./{year}/Divvy_Trips_{year}_{quarter}.csv')
    elif year == '2020':
        for quarter in quarters[:1]:
            quarter_file_paths.append(f'./{year}/Divvy_Trips_{year}_{quarter}.csv')
    else:
        for quarter in quarters:
            quarter_file_paths.append(f'./{year}/Divvy_Trips_{year}_{quarter}.csv')



```

```python
column_name_by_year = {
    '2016': 'starttime',
    '2017': 'start_time',
    '2018': 'start_time',
    '2019': 'start_time',
    '2020': 'started_at'
}
```

```python
daily_counts = []


for file in quarter_file_paths:
    df = pd.read_csv(file)
    year = file.split('/')[1]
    quarter = file.split('_')[-1].split('.')[0]

    # Handle mismatching column names
    if year == '2018' and quarter == 'Q1':
        df["start_time"] = None
        df['start_time'] = df['01 - Rental Details Local Start Time']
        df["usertype"] = None
        df['usertype'] = df['User Type']
        df['trip_id'] = df['01 - Rental Details Rental ID']
    if year == '2019' and quarter == 'Q2':
        df["start_time"] = None
        df['start_time'] = df['01 - Rental Details Local Start Time']
        df["usertype"] = None
        df['usertype'] = df['User Type']
        df['trip_id'] = df['01 - Rental Details Rental ID']
    else:
        df["start_time"] = df[column_name_by_year[year]]

    if year == '2020':
        df['usertype'] = df['member_casual']
        df['trip_id'] = None
        df['trip_id'] = df['ride_id']
        
    df["Year"] = pd.to_datetime(df["start_time"]).dt.year
    df["Month"] = pd.to_datetime(df["start_time"]).dt.month
    df["Day"] = pd.to_datetime(df["start_time"]).dt.day

    print(file)
    # Aggregate data
    grouped = df.groupby(["Year", "Month", "Day"]).agg(
        **{"Total Counts": ("trip_id", "count"),
        "Casual User Count": ("usertype", lambda x: (x == "Customer").sum()),
        "Member User Count": ("usertype", lambda x: (x == "Subscriber").sum())}
    ).reset_index()

    grouped["Classic Bike Count"] = 0
    grouped["Electric Bike Count"] = 0

    daily_counts.append(grouped)

# Combine all months' data into one DataFrame
daily_counts_df_Q = pd.concat(daily_counts, ignore_index=True)

daily_counts_df_Q = daily_counts_df_Q.groupby(["Year", "Month", "Day"], as_index=False).sum()

daily_counts_df_Q["Date"] = pd.to_datetime(daily_counts_df_Q["Year"].astype(int).astype(str) + "-" + daily_counts_df_Q["Month"].astype(str) + "-" + daily_counts_df_Q["Day"].astype(str))

daily_counts_df_Q["Week"] = daily_counts_df_Q["Date"].dt.to_period("W-SUN").apply(lambda r: r.start_time)

daily_df_Q = daily_counts_df_Q.drop(columns=['Year', 'Month', 'Day', 'Date'])

# Group by "Week" and sum up the counts
weekly_counts_df_Q = daily_df_Q.groupby("Week", as_index=False).sum()
```

```python
# Handling file path for monthly reports from 2020 - 2025 February
years = ['2020', '2021', '2022', '2023', '2024', '2025']
file_paths = []
for yr in years:
    if yr == '2025':
        yr_file_paths = [f"./{yr}/{yr}0{m}-divvy-tripdata.csv" for m in range(1, 3)]
    elif yr == '2020':
        yr_file_paths = [f"./{yr}/{yr}0{m}-divvy-tripdata.csv" for m in range(4, 10)] + [f"./{yr}/{yr}{m}-divvy-tripdata.csv" for m in range(10, 13)]
    else:
        yr_file_paths = [f"./{yr}/{yr}0{m}-divvy-tripdata.csv" for m in range(1, 10)] + [f"./{yr}/{yr}{m}-divvy-tripdata.csv" for m in range(10, 13)]
    file_paths += yr_file_paths

```

```python
daily_counts = []


for file in file_paths:
    df = pd.read_csv(file)

    df["Year"] = pd.to_datetime(df["started_at"]).dt.year
    df["Month"] = pd.to_datetime(df["started_at"]).dt.month
    df["Day"] = pd.to_datetime(df["started_at"]).dt.day


    # Aggregate data
    grouped = df.groupby(["Year", "Month", "Day"]).agg(
        **{"Total Counts": ("ride_id", "count"),
        "Casual User Count": ("member_casual", lambda x: (x == "casual").sum()),
        "Member User Count": ("member_casual", lambda x: (x == "member").sum()),
        "Classic Bike Count": ("rideable_type", lambda x: (x == "classic_bike").sum()),
        "Electric Bike Count": ("rideable_type", lambda x: (x == "electric_bike").sum())}
    ).reset_index()

    daily_counts.append(grouped)

# Combine all months' data into one DataFrame
daily_counts_df = pd.concat(daily_counts, ignore_index=True)

daily_counts_df = daily_counts_df.groupby(["Year", "Month", "Day"], as_index=False).sum()

daily_counts_df["Date"] = pd.to_datetime(daily_counts_df["Year"].astype(int).astype(str) + "-" + daily_counts_df["Month"].astype(str) + "-" + daily_counts_df["Day"].astype(str))

daily_counts_df["Week"] = daily_counts_df["Date"].dt.to_period("W-SUN").apply(lambda r: r.start_time)

daily_df = daily_counts_df.drop(columns=['Year', 'Month', 'Day', 'Date'])

# Group by "Week" and sum up the counts
weekly_counts_df = daily_df.groupby("Week", as_index=False).sum()
```
