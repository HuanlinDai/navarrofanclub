```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
from pmdarima import auto_arima


file_path = "/content/Divvy Time Series Dataset - Sheet1.csv"  # Change path if needed
df = pd.read_csv(file_path)
df["Year"] = df["Year"].fillna(method='ffill').fillna(method='bfill')
df["Docked Bike Count"] = df["Docked Bike Count"].fillna(method='ffill').fillna(method='bfill')


df["Date"] = pd.to_datetime(df[['Year', 'Month']].assign(day=1))
df.set_index("Date", inplace=True)


df = df.asfreq('MS').sort_index()

df_time_series = df[["Docked Bike Count"]].copy()


plt.figure(figsize=(10, 5))
plt.plot(df_time_series.index, df_time_series["Docked Bike Count"], marker='o', linestyle='-', color='orange', label="Observed")
plt.xlabel("Date")
plt.ylabel("Docked Bike Count")
plt.title("Docked Bike Count Over Time")
plt.legend()
plt.grid(True)
plt.show()


auto_model = auto_arima(df_time_series["Docked Bike Count"].dropna(),
                        seasonal=True, m=12, trace=True, stepwise=True,
                        max_p=3, max_q=3, max_P=3, max_Q=3, max_d=2, max_D=2)

print(auto_model.summary())


sarimax_model = SARIMAX(df_time_series["Docked Bike Count"],
                        order=auto_model.order,
                        seasonal_order=auto_model.seasonal_order,
                        enforce_stationarity=False, enforce_invertibility=False)
sarimax_results = sarimax_model.fit()

df_time_series["SARIMAX_Predicted"] = sarimax_results.fittedvalues
forecast_steps = 12
sarimax_forecast = sarimax_results.get_forecast(steps=forecast_steps)
sarimax_forecast_index = pd.date_range(start=df_time_series.index[-1] + pd.DateOffset(months=1), periods=forecast_steps, freq='MS')
sarimax_forecast_values = sarimax_forecast.predicted_mean

df_prophet = df_time_series.reset_index()[['Date', 'Docked Bike Count']]
df_prophet.columns = ['ds', 'y']

prophet_model = Prophet()
prophet_model.fit(df_prophet)

future_dates = prophet_model.make_future_dataframe(periods=12, freq='MS')

prophet_forecast = prophet_model.predict(future_dates)

df_comparison = df_time_series[['Docked Bike Count', 'SARIMAX_Predicted']].dropna()

plt.figure(figsize=(12, 6))
plt.plot(df_time_series.index, df_time_series["Docked Bike Count"], marker='o', linestyle='-', color="orange", label="Observed")
plt.plot(sarimax_forecast_index, sarimax_forecast_values, marker='o', linestyle='--', color='red', label="SARIMAX Forecast")
plt.plot(prophet_forecast['ds'], prophet_forecast['yhat'], linestyle='--', color='blue', label="Prophet Forecast")
plt.xlabel("Date")
plt.ylabel("Docked Bike Count")
plt.title("SARIMAX (Auto-ARIMA) vs. Prophet Forecast for Docked Bike Count")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df_comparison.index, df_comparison["Docked Bike Count"], marker='o', linestyle='-', color="orange", label="Actual")
plt.plot(df_comparison.index, df_comparison["SARIMAX_Predicted"], linestyle='--', color='red', label="SARIMAX Predicted")
plt.xlabel("Date")
plt.ylabel("Docked Bike Count")
plt.title("Actual vs. Predicted (SARIMAX Auto-ARIMA)")
plt.legend()
plt.grid(True)
plt.show()
```

The SARIMAX model demonstrates strong short-term predictive accuracy, closely aligning with historical data, particularly during the sharp peaks in 2020-2021. However, its long-term forecast appears overly stable, potentially underfitting trend changes. In contrast, the Prophet model captures seasonal fluctuations but introduces excessive volatility, sometimes predicting negative docked bike counts, which is unrealistic. SARIMAX’s conservative approach makes it reliable for steady patterns, while Prophet’s responsiveness may be useful for capturing seasonality, though it requires tuning to prevent erratic swings. To improve accuracy, SARIMAX could benefit from higher seasonal AR or MA terms and external regressors, while Prophet should adjust changepoint_prior_scale and implement floor constraints. A potential next step is exploring an ensemble approach that balances both models’ strengths for more robust forecasting.
