```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from pmdarima import auto_arima

# Load dataset
df = pd.read_csv("Divvy Time Series Dataset - Sheet1.csv")

# Ensure Month is numeric
df['Month'] = pd.to_numeric(df['Month'], errors='coerce')

# Drop any rows where Month is missing
df.dropna(subset=['Month'], inplace=True)

# Convert Month to integer
df['Month'] = df['Month'].astype(int)

# Create a sequential time index
df['TimeIndex'] = range(1, len(df) + 1)
df.set_index('TimeIndex', inplace=True)

# Extract the time series
ts = df['Member User Count']

# Automatically determine SARIMA order
s_model = auto_arima(ts, seasonal=True, m=12, stepwise=True, trace=True)

# Fit the best SARIMA model
sarima_model = sm.tsa.SARIMAX(ts, order=s_model.order, seasonal_order=s_model.seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
results = sarima_model.fit()

# Print model summary
print(results.summary())

# Forecast for the next 12 months
forecast = results.get_forecast(steps=12)
forecast_index = np.arange(ts.index[-1] + 1, ts.index[-1] + 13)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(ts, label="Observed", color='blue')
plt.plot(forecast_index, forecast_mean, label="Forecast", color='red')
plt.fill_between(forecast_index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink', alpha=0.3)
plt.xlabel("Time Index")
plt.ylabel("Member User Count")
plt.title("SARIMA Forecast of Member User Count")
plt.legend()
plt.grid()
plt.show()
```
